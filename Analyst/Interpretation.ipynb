```python
# ============================================================
# 04_FuzzyScore_Interpretation.ipynb
# Analisis Distribusi dan Interpretasi Fuzzy Score
# Untuk Bab 4  Subbab 4.4.5
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 1. Load dataset hasil fuzzy grading
# ------------------------------------------------------------
path = r"E:\DragonEye\dataset\graded_features.csv"
df = pd.read_csv(path)

print("Jumlah data:", len(df))
df.head()

# ------------------------------------------------------------
# 2. Tabel Ringkasan 15 Sampel
# ------------------------------------------------------------

sample_df = df[['length_cm','diameter_cm','weight_est_g',
                'ratio','fuzzy_score','final_grade']].head(15)

sample_df


# ------------------------------------------------------------
# 3. Histogram Distribusi Fuzzy Score
# ------------------------------------------------------------

plt.figure(figsize=(10,6))
plt.hist(df['fuzzy_score'], bins=20)
plt.title("Histogram Distribusi Fuzzy Score")
plt.xlabel("Fuzzy Score (0100)")
plt.ylabel("Frekuensi")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 4. Scatter Plot: Bobot vs Fuzzy Score
# ------------------------------------------------------------

plt.figure(figsize=(10,6))
plt.scatter(df['weight_est_g'], df['fuzzy_score'])
plt.title("Hubungan Berat Buah dan Fuzzy Score")
plt.xlabel("Berat (g)")
plt.ylabel("Fuzzy Score")
plt.grid(True)
plt.show()


# ------------------------------------------------------------
# 5. Scatter Plot: Rasio vs Fuzzy Score
# ------------------------------------------------------------

plt.figure(figsize=(10,6))
plt.scatter(df['ratio'], df['fuzzy_score'])
plt.title("Hubungan Rasio Bentuk (Panjang/Diameter) dan Fuzzy Score")
plt.xlabel("Rasio (Panjang / Diameter)")
plt.ylabel("Fuzzy Score")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 6. Distribusi Frekuensi Final Grade
# ------------------------------------------------------------

print("Distribusi Final Grade:")
print(df['final_grade'].value_counts())

df['final_grade'].value_counts().plot(kind='bar', figsize=(7,5))
plt.title("Distribusi Final Grade Berdasarkan Standar Industri")
plt.xlabel("Grade")
plt.ylabel("Jumlah")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 7. Kesimpulan Singkat (untuk dicopy ke laporan)
# ------------------------------------------------------------

print("""
Interpretasi Awal:
- Fuzzy score meningkat seiring bertambahnya panjang, diameter, dan berat.
- Buah dengan rasio bentuk kurang ideal cenderung memiliki fuzzy score lebih rendah.
- Distribusi fuzzy score menunjukkan pola yang konsisten dan tidak ada anomali ekstrem.
Silakan gunakan hasil tabel dan grafik ini untuk subbab 4.4.5.
""")
```
